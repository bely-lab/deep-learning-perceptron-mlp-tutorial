#I. **Perceptron**

**Overview**

The Perceptron, pioneered by Frank Rosenblatt, represents a fundamental advancement in neural network models. Unlike the McCulloch-Pitts neuron model, which provided a binary output, Rosenblatt's Perceptron introduced the concept of adjustable weights, transforming it into a linear binary classifier specifically designed for supervised learning tasks. Rosenblatt ingeniously merged principles from the McCulloch-Pitts neuron model and the Hebbian learning rule, resulting in a more adaptable and capable neural network architecture.

**Key Innovations of the Perceptron**

Rosenblatt's Perceptron introduced several key innovations:

**A. Variable Weight Values:**

Unlike the fixed weights in the McCulloch-Pitts model, the Perceptron incorporates variable weight values. This flexibility allows the model to learn and adapt to complex patterns within the data.

**B. Bias Input:**

In addition to the input features, the Perceptron includes an extra input representing bias. This bias term enables the Perceptron to better capture and model real-world data.

**C. Autonomous Learning:**

Rosenblatt proposed that artificial neurons could autonomously learn optimal weightings and thresholds from training data. This departure from fixed parameters laid the foundation for a more dynamic and adaptable learning process





**1. Perceptron Model**

The perceptron is a fundamental building block of neural networks. It takes inputs, applies weights to them, sums them up, and passes them through an activation function to produce an output.

 **2. Define the Perceptron Class Training Algorithm (Perceptron Learning Rule)**

 The perceptron learning rule adjusts the weights based on the error in prediction. It updates the weights and bias iteratively to minimize the error.

```python
class Perceptron:
    def __init__(self, input_size, learning_rate=0.01):
        self.weights = np.zeros(input_size + 1)  # +1 for bias
        self.learning_rate = learning_rate

    def activate(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        # Add bias term (x_0 = 1)
        x_with_bias = np.insert(x, 0, 1)
        return self.activate(np.dot(self.weights, x_with_bias))

    def train(self, X, y, epochs):
        for _ in range(epochs):
            for i in range(len(y)):
                # Add bias term (x_0 = 1)
                x_with_bias = np.insert(X[i], 0, 1)
                prediction = self.predict(X[i])
                error = y[i] - prediction
                self.weights += self.learning_rate * error * x_with_bias
```

**Applications:**

**A.Implementing AND Gate**







```python
# Example: AND Gate
# Prepare Training Data
X_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])


# Create and Train the Perceptron
perceptron_and = Perceptron(input_size=2)
perceptron_and.train(X_and, y_and, epochs=10)

# Test the Perceptron and Print the result
print("AND Gate Predictions:")
for input in X_and:
    print(f"Input: {input}, Output: {perceptron_and.predict(input)}")
```

**B.Implementing XOR**

```python
# Example: XOR Gate
# Prepare Training Data
X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])

# Create and Train the Perceptron
perceptron_xor = Perceptron(input_size=2)
perceptron_xor.train(X_xor, y_xor, epochs=10)

# Test the Perceptron and Print the result
print("\nXOR Gate Predictions:")
for input in X_xor:
    print(f"Input: {input}, Output: {perceptron_xor.predict(input)}")
```

# **2. Data representation in Keras**


**Keras** uses a multidimensional array of Numpy as a basic data structure, calls this data structure a tensor. In short, we could say that a tensor has three main attributes:
1.   Number of axes (Rank)

**- 0-Dimensional Tensor (Scalar):**

* A single numerical value is a scalar tensor. It has no dimensions and cannot be further decomposed into smaller arrays.

* In Python NumPy library, we can create a scalar using a single numerical value or functions like **np.array(123)**.

* The **ndim** attribute of a scalar tensor is 0.

**- 1-Dimensional Tensor (Vector):**

* An array of numbers arranged in a single row or column is a vector tensor.
It has one dimension.

* In NumPy, we can create a vector using a list or array literal, for example: **np.array([1, 2, 3])**.

* The ndim attribute of a 1D tensor is 1.

**- 2-Dimensional Tensor (Matrix):**

* An array of vectors (rows) arranged in a rectangular grid is a matrix tensor. It has two dimensions: rows and columns.

* In NumPy, we can create a matrix using nested lists or array literals. For instance: **np.array([[1, 2, 3], [4, 5, 6]])**.

* The ndim attribute of a 2D tensor is 2.

**- Higher-Dimensional Tensors:**

* Tensors can extend to more than two dimensions. For example, a 3D tensor (sometimes visualized as a cube) has three dimensions, a 4D tensor has four, and so on.

* Higher-dimensional tensors are essential in various machine learning and deep learning applications.

* In NumPy, you can create tensors with higher dimensions using nested list/array structures.

*Note*: In the Python Numpy library this is called the tensor **ndim**

2.   Shape
* The shape of a tensor is a tuple of integers that specifies the number of elements along each dimension (axis) of the tensor.

* It essentially defines the size and structure of the data contained within the tensor.
* Each integer in the shape tuple represents the length of that particular dimension. For example, a shape of (2, 3, 4) indicates a 3D tensor with 2 elements in the first dimension, 3 elements in the second dimension, and 4 elements in the third dimension.

* In the NumPy library, the shape attribute of a tensor is directly accessible using the **.shape** property. It returns a tuple of integers representing the dimensions of the tensor.

3.   Data type

* The data type of a tensor specifies the kind of numerical data each element within the tensor holds.
* It determines how the numbers are represented and stored in memory, which influences calculations and memory usage
* Example **uint8, float32, float64, etc**
* In the Numpy library this attribute is called **dtype**.

# **3. Data Preprocessing**

**Transformation:**

* Transformation of the tensor (image) from 2 dimensions (2D) to a vector of 1 dimension (1D) is required
* the matrix of 28*28 numbers can be represented by a vector (array) of 784 numbers (concatenating row by row).
* In Python, converting every image of the MNIST dataset to avector with 784 components can be accomplished as follows:

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

**Image Scaling and Data Normalization**

* Original Data: Pixel values stored as unsigned 8-bit integers** (uint8)** ranging from 0 (black) to 255 (white).
* Normalizing input data to a specific range (e.g., [0, 1]) for better neural network training.
* Standardizes the range of features (pixel values).
* Facilitates optimization for certain activation functions.


```python
# normalize the input images
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

```python
# verify that x_train.shape and x_test.shape

print(x_train.shape)

print(x_test.shape)
```

**Transforming the labels into a vector**

```python
from keras.utils import to_categorical

# befor transforming the lables to a vector
print("# befor transforming the lables to a vector")
print(y_train[0])

print(y_test[0])

print(y_train.shape)

print(y_test.shape)


# After transforming the lables to a vector
print("After transforming the lables to a vector")
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

print(y_test[0])

print(y_train[0])

print(y_train.shape)

print(y_test.shape)

```

# **4. Define the Model Architecture**

**Sequential class in Keras**
* The main data structure in Keras, which allows the creation of a basic neural network
* Keras also offers an API that allows implementing more complex models in the form of a graph that can have multiple inputs, multiple outputs, with arbitrary connections in between
* The **Sequential** class of the Keras library is a wrapper for the sequential neural network model that Keras offers and can be created in the following way:

```python
from keras.models import Sequential
model = Sequential()
```

In Keras  all the required types of layers can be easily added to the model through the **add()** method

**Defining the model**

```python
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential()
model.add(Dense(10, activation='sigmoid', input_shape=(784,)))
model.add(Dense(10, activation='softmax'))
```

The neural network has been defined as a sequence of two layers that are densely connected (or fully connected), meaning that all the neurons in each layer are connected to all the neurons in the next layer.

**Input Layer:**

* Explicitly define the input data shape using the **input_shape** argument.
In this example, the input is a tensor with shape (None, 784), indicating:
- None: The batch size can be variable .
- 784: The number of features (784 pixels in a 28x28 image).

**Keras Automatic Shape Inference:**

- Keras automatically infers the shape of tensors between layers after the first layer.
- This simplifies code by only requiring the programmer to specify the input shape.

**Layers:**

Each layer definition includes:

- Number of nodes (neurons) in the layer.

- Activation function applied to the layer's outputs.

**Output Layer (Softmax):**

The second layer is a softmax layer with 10 neurons.
- It outputs a probability vector of size 10:
- Each element represents the probability of the input image belonging to a specific digit (0-9).
- This is typical for multi-class classification (10 output neurons for 10 classes).
- Binary classification would only require one neuron for a single probability value.

# **5. Compile the Model**

Compile the model by specifying the loss function, optimizer, and evaluation metric.

```python
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics = ['accuracy'])
```

# **6. Train the Model**

Once our model has been defined and the learning method configured, it is ready to be trained.

```python
#model.fit(x_train, y_train, batch_size=100, epochs=5)
history = model.fit(x_train, y_train, epochs=50, batch_size=100, validation_split=0.2)
```

#  **7. Model evaluation**

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print('Test accuracy:', test_accuracy)

```

# **8. Visualize Training History**

Visualize the training and validation accuracy and loss over epochs to understand the model's training behavior.

```python
import matplotlib.pyplot as plt

history_dict = history.history
accuracy = history_dict['accuracy']
val_accuracy = history_dict['val_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(accuracy) + 1)

plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
